#**Extração de dados via API**

```python
#Import das libs utilizadas
import matplotlib.pyplot as plt
import pandas as pd
import requests
```

```python
url = 'https://ghibliapi.herokuapp.com/films'
```

```python
#Chamada da API 
r = requests.get(url)
r
```

```python
#Extrai os dados .json
json = r.json()
```

```python
json
```

#**Transformação dos dados**

```python
#Cria o DataFrame 
df = pd.DataFrame(json)
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df.head(1)
```

```python
df = df.rename(columns={"title" : "titulo", 
                        "original_title" : "titulo_japones", 
                        "original_title_romanised" : "titulo_original_romanizado",
                        "image" : "imagem",
                        "movie_banner" : "banner",
                        "description" : "descricao",
                        "director" : "diretor",
                        "producer" : "produtor", 
                        "release_date" : "data_lancamento",
                        "running_time" : "duracao",
                        "rt_score" : "nota"})
```

```python
retirar_colunas = df.loc[:, ["imagem", "banner", "descricao", "people", "species", "locations", "vehicles", "url"]]
df.drop(retirar_colunas, inplace=True, axis=1)
```

```python
df.loc[1:3]
```

#**Análise Exploratória**

```python
filmes_por_ano = df.loc[:, ["data_lancamento", "titulo","nota"]]
filmes_por_ano.head(3)
```

```python
filmes_por_ano["nota"] = filmes_por_ano["nota"].astype('int32')
```

```python
filmes_por_ano.groupby("data_lancamento")["titulo"].count().sort_values(ascending=True).plot.bar(title="Filmes por Ano")
plt.xlabel("Ano")
plt.ylabel("Quantidade");
```

```python
filmes_por_ano.describe()
```

```python
filmes_por_ano.groupby("data_lancamento")["nota"].max().plot(title="Variação de notas por ano")
plt.xlabel("Ano de Lançamento")
plt.ylabel("Notas");
```
